```python
questions = survey_data[['question_1_collector_description',
       'question_2_collector_goal', 'question_3a_grading_interests',
       'question_3b_grading_interests', 'question_3c_grading_interests',
       'question_3d_grading_interests', 'question_3e_grading_interests',
       'question_3f_grading_interests', 'question_3g_grading_interests',
       'question_4a_topic_interests', 'question_4b_topic_interests',
       'question_4c_topic_interests', 'question_4d_topic_interests',
       'question_4e_topic_interests', 'question_4f_topic_interests',
       'question_4g_topic_interests']].reset_index(drop=True)

for col in questions.columns:
    print(questions[col].unique())
```

```python
for col in questions.columns:
    print(questions[col].count())
```

```python
questions = questions.fillna('missing')
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=True) 

one_hot_encoded = encoder.fit_transform(questions)

one_hot_df = pd.DataFrame.sparse.from_spmatrix(one_hot_encoded, columns=encoder.get_feature_names_out())

one_hot_df.head()
```

```python
survey_meta_data = survey_data.drop(questions, axis=1)

survey_meta_data.head(5)
```

```python
survey_data_encoded = pd.concat([survey_meta_data, one_hot_df], axis=1)

survey_data_encoded.head(5)
```
